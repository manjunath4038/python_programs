```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class Linkedlist:
    def __init__(self):
        self.head = None
        
    def printList(self):
        temp = self.head
        print("printing linked list")
        if temp is None:
            print("List is Empty")
            return
        while temp :
            print(temp.data)
            temp = temp.next
    
    def insert_at_the_front(self,new_data):
        print("Inserting at the front", new_data)
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node
    
    def insert_at_the_end(self,new_data):
        new_node = Node(new_data)
        print("Inserting at the end",new_data)
        if(self.head is None):
            self.head = new_node
            return
        last = self.head
        while (last.next):
            last = last.next
        last.next = new_node
    
    def insert_after_node(self,prev_node,new_data):
        print("Inserting at the after",prev_node.data,"given node", new_data)
        new_node = Node(new_data)
        new_node.next = prev_node.next
        prev_node.next = new_node
        
    def delete_node(self,key):
        print("Deleting a node with the key", key)
        temp = self.head
        # If head has the key to be deleted
        if temp is not None:
            if temp.data == key:
                self.head = temp.next
                temp = None
                return
        while(temp is not None):
            if (temp.data == key):
                break
            prev = temp
            temp = temp.next
        if(temp == None):
            print("Key is not found")
            return
        prev.next = temp.next
        temp = None
        
    def delete_node_from_position(self,position):
        print("Deleting a node from the position", position)
        
        # If the list is empty
        if (self.head is None):
            return
        
        # Store the head node
        temp = self.head
        
        # If head needs to be removed
        if (position == 0):
            self.head = temp.next
            temp = None
            
        # Find previous node of the node to be deleted
        for i in range(position -1):
            temp = temp.next
            if temp is None:
                break
        ## If position is more than number of nodes        
        if temp is None:
            return
        if temp.next is None:
            return
        
        # Node temp.next is the node to be deleted 
        # store pointer to the next of node to be deleted 
        next = temp.next.next
        
        # Unlink the node from linked list
        temp.next = None
        temp.next = next
        
    def delete_entire_list(self):
        print("Deleting the entire list")
        
        current = self.head
        
        while current:
            prev = current.next
            del current.data
            current = prev
            
    def find_length_iterative(self):
        print("Finding length of the list using itervative method")
        curr = self.head
        count = 0
        if curr is not None:
            while curr:
                curr = curr.next
                count += 1
        return count
    
    def get_count_rec(self, node):
        if not node:
            return 0
        
        return 1 + self.get_count_rec(node.next)
        
    def find_length_recursive(self):
        print("Finding length of the list using Recursive method")
        return self.get_count_rec(self.head)
    
    def search_element_iter(self, key):
        current = self.head
        while current:
            if current.data == key:
                print("Key is found", key)
                break
            current = current.next
        
        if current is None:
            print("Key is not found")
            
    def search_element_rec(self, node, key):
        if node is None:
            return False
        if node.data == key:
            return True
        return self.search_element_rec(node.next, key)
    
    def reverse_list(self):
        print("reserve the linked list")
        prev = None
        cur = self.head
        
        while cur is not None:
            next = cur.next
            cur.next = prev
            prev = cur
            cur = next
        self.head = prev
    
    def middle_of_list(self):
        
        slow_pointer = self.head
        fast_pointer = self.head
        
        while(fast_pointer is not None and fast_pointer.next is not None):
            slow_pointer = slow_pointer.next
            fast_pointer = fast_pointer.next.next
        print("Middle of the list",slow_pointer.data)
```
